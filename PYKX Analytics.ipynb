### qsql in function form

```python
# select

tab_btc.select(columns = ['close', 'open'])

#tab_btc.select(columns = kx.Column('open')) # bit irritation needs to choose column one by one
```

```python
# where condition

tab_btc.select(columns = ['timestamp','open', 'close'], where = kx.Column('timestamp').date == kx.DateAtom(2017,8,23))


# ahh long function query
```

```python
tab_btc.select(columns = ['timestamp','open', 'close'], where = ((kx.Column('timestamp').date == kx.DateAtom(2017,8,23)) & (kx.Column('open') >= kx.RealAtom(4100))))
```

```python
# by clause

tab_btc.select(columns = kx.Column('open').last()& kx.Column('close').max(),by=kx.Column('timestamp').date)
```
